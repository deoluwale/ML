DATA CLEANING

```python
##Import pandas and numpy
import pandas as pd
import numpy as np

##Read in the dataset
df = pd.DataFrame({'From_To': ['LoNDon_paris', 'MAdrid_miLAN',
'londON_StockhOlm',

'Budapest_PaRis', 'Brussels_londOn'],
'FlightNumber': [10045, np.nan, 10065, np.nan, 10085],
'RecentDelays': [[23, 47], [], [24, 43, 87], [13], [67, 32]],
'Airline': ['KLM(!)', '<Air France> (12)', '(British Airways. )',
'12. Air France', '"Swiss Air"']})
df
```

```python
##1
df['FlightNumber'].fillna(value= 10055, inplace=True, limit=1)
df['FlightNumber'].fillna(value= 10075, inplace=True, limit=1)

##convert the FlightNumber column to integer 
df['FlightNumber'] = df['FlightNumber'].astype(int)
df
```

```python
##2
##split the From-To column to give a temporary dataframe
temp_df = df["From_To"].str.split("_", expand = True, )
temp_df = temp_df.rename(columns={0: 'From', 1:'To'}, )
temp_df
```

```python
##3
##Standardize the datapoints to title case
temp_df = temp_df.apply(lambda x: x.astype(str).str.title())
temp_df
```

```python
##4
##Delete the From_To column from df
df.drop(columns=['From_To'], axis=1, inplace=True)
df = pd.concat([temp_df,df], axis=1)
df
```

```python
##5
##Making Delays dataframe
delays = pd.DataFrame(df.RecentDelays.tolist(), ).rename(columns={0:'Delay_1', 1:'Delay_2', 2:'Delay_3'})
df.drop(['RecentDelays'], axis=1, inplace=True)
df = pd.concat([df,delays], axis=1)
df
```
